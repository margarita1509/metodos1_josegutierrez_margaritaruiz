```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
f1 = lambda x: 2*x-2
f2 = lambda x: (1-x)/2
f3 = lambda x: 4-x
```

```python
x0 = np.arange(20)

y1 = f1(x0)
y2 = f2(x0)
y3 = f3(x0)
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
b = np.array([2,1,4])

AT = A.T
prod = AT@A
pt = np.linalg.inv(prod)@AT@b

print('Se observa que el punto común entre las líneas es',pt)
```

```python
plt.plot(x0,y1)
plt.plot(x0,y2)
plt.plot(x0,y3)
plt.scatter(pt[0],pt[1],color='red')
```

```python
def f(x,y):
    a1 = 2*x-2-y
    a2 = x+2*y-1
    a3 = x+y-4
    return np.sqrt(a1**2 + a2**2 + a3**2)
f = np.vectorize(f)
```

```python
X = np.arange(-5, 5, 0.01)
Y = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(X, Y)

Z = f(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='gist_rainbow')
ax.view_init(15,20)
ax.scatter(pt[0],pt[1],f(pt[0],pt[1]),c='b')
```

```python
print('Se observa que al graficar en R3 la distancia mínima corresponde al punto más bajo de la superficie,\nlo cual es el valor de la coordenada z y demuestra el punto más cercano entre líneas que no tienen punto común.\n')
```
