```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import sympy as sym
from scipy.integrate import quad
from scipy.integrate import dblquad
```

```python
n = 200
x = np.linspace(-1,1,n+1)
y = np.linspace(-1,1,n+1)
```

```python
def f(x,y,R=1.):
    
    z = R**2 - x**2 - y**2
    
    if z <= 0.:
        return 0.
    else:
        return np.sqrt(z)
```

```python
f = np.vectorize(f)
```

```python
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='coolwarm')
```

```python
vol = 0
for i in range(len(Z)-1):
    for j in range(len(Z)-1):
        p1 = Z[i][j]
        p2 = Z[i][j+1]
        p3 = Z[i+1][j]
        p4 = Z[i+1][j+1]
        P = (p1+p2+p3+p4)/40000
        vol += P
        
print(vol)
```
