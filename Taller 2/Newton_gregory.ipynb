```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([0,1.5,2,3.8,4.2,5.9])
Y=np.array([-18,-13,0,5,3,10])
```

```python
def interpol_NG(X,Y,x_ev):
    C=np.zeros([len(X),len(Y)])
    C[:,0]=Y
    for i in range(1,len(X)):
        for j in range(len(X)-i):
            C[j][i]=(C[j+1][i-1]-C[j][i-1])/(X[j+i]-X[j])
    dif_div=C[0]
    pol=0
    for i in range(len(X)):
        térm=dif_div[i]
        for j in range(i):
            térm*=(x_ev-X[j])
        pol+=térm
    return pol
```

```python
ejex=np.linspace(X[0],X[-1],100)
ejey=[]
for i in range(len(ejex)):
    ejey.append(interpol_NG(X,Y,ejex[i]))
plt.scatter(X,Y,color='r',marker='o')
plt.plot(ejex,ejey)
```
