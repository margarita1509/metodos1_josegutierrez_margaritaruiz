```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problema 12 (ensamble microcanónico)
#d
def Entropía(x):
    S=-1*(x*np.log(x)+(1-x)*np.log(1-x))
    return S
entro=np.zeros(100)
x=np.linspace(0,1,100)
for i in range(len(x)):
    if x[i]==1 or x[i]==0:
        entro[i]=0
    else:
        entro[i]=Entropía(x[i])
plt.plot(x,entro,color='r')
plt.xlabel('x')
plt.ylabel('Entropía')
plt.grid()
plt.show()
```
