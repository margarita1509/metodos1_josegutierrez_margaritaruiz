```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from math import factorial as fac
```

```python
#Problema 4 (cumple)
#Forma general del cálculo simplificada
def prob_dif_CUM(n):
    num=fac(365)/fac(365-n)
    P_dif_cum=num/(365**n)
    return P_dif_cum
#Grafica para n=80
prob_n=np.zeros(80)
num_per=np.zeros(80)
for n in range(len(prob_n)):
    prob_n[n]=prob_dif_CUM(n+1)
    num_per[n]=n
plt.plot(num_per,prob_n,color='r')
plt.xlabel('Número de personas')
plt.ylabel('Probabilidad de no repetir cum')
plt.grid()
plt.show()
```
