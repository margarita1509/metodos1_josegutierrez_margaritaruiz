```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from mpl_toolkits.mplot3d import axes3d
```

```python
#Problema 9 (monedas truncadas)
p1=np.linspace(0.1,0.9,100)
p2=np.linspace(0.1,0.5,100)
def P_caraysello(p1,p2):
    P_carasello=0.5*(p1*(1-p2)+p2*(1-p1))+0.25*(p1*p2+(1-p2)*(1-p1))
    return P_carasello
#graficar
fig=plt.figure()
X,Y=np.meshgrid(p2,p1)
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,P_caraysello(X,Y),cmap='inferno',alpha=0.6)
ax.set_xlabel('p2')
ax.set_ylabel('p1')
ax.set_zlabel('Prob Cara y sello')
plt.show()
#Máximos y Mínimos
maximo=0
minimo=1
max_cord=np.zeros(2)
min_cord=np.zeros(2)
for i in range(len(p1)):
    for j in range(len(p2)):
        if P_caraysello(p1[i],p2[j])>maximo:
            max_cord[0]=p1[i]
            max_cord[1]=p2[j]
            maximo=P_caraysello(p1[i],p2[j])
        if P_caraysello(p1,p2)[i]<minimo:
            min_cord[0]=p1[i]
            min_cord[1]=p2[j]
            minimo=P_caraysello(p1[i],p2[j])
minimo,min_cord,maximo,max_cord
```
