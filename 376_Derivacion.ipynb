```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
#f = Function(x,y)
```

```python
def Function(x:np.ndarray,y:np.ndarray):
    v = 2
    r = 2 
    for i in range(25):
        for j in range(25):
            if (x)**2+(y)**2<=4:
                flujo = 0
            else:
                flujo = v*x*(1-((r**2)/((x**2)+(y**2))))
    return flujo
```

```python

h = 0.001 
vx = np.zeros((25,25))
vy = np.zeros((25,25))
for i in range(0,25):
    for j in range(0,25):
        if (x[i])**2+(y[j])**2<=4:
            vx[i][j]=(0)
            vy[i][j]=(0)
        else:
            vx[i][j] = (Function(x[i]+h,y[j]) - Function(x[i]-h,y[j]))/(2*h)
            vy[i][j] = -1*(Function(x[i],y[j]+h) - Function(x[i],y[j]-h))/(2*h)
            
   
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
circle = plt.Circle((0, 0), 2, color='r', fill=False)
ax.add_patch(circle)
    
for i in range(25):
    for j in range(25):
        ax.quiver(x[i],y[j],vx[i][j], vy[i][j])
```
